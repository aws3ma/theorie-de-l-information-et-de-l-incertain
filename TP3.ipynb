```python
import seaborn as sns
data = sns.load_dataset("iris")
```

```python
print(data.head()) #afficher 5 premier lignes , data ayant 5 features
```

```python
sep_len = data["sepal_length"]
print(sep_len)
```

```python
sep_wid = data["sepal_width"]
print(sep_wid)
```

```python
pet_len = data["petal_length"]
print(pet_len)
```

```python
pet_wid = data["petal_width"]
print(pet_wid)
```

```python
n=50
sep_lenc1 =sep_len[0:n]
sep_lenc2 =sep_len[n:n+50]
sep_lenc3 =sep_len[n+50:50+2*n]
```

```python
n=50
sep_widc1 =sep_wid[0:n]
sep_widc2 =sep_wid[n:n+50]
sep_widc3 =sep_wid[n+50:50+2*n]
```

```python
n=50
pet_lenc1 =pet_len[0:n]
pet_lenc2 =pet_len[n:n+50]
pet_lenc3 =pet_len[n+50:50+2*n]
```

```python
n=50
pet_widc1 =pet_wid[0:n]
pet_widc2 =pet_wid[n:n+50]
pet_widc3 =pet_wid[n+50:50+2*n]
```

```python
from scipy.stats import skew,kurtosis #kutosis applatissement, skew coeff
desc=sep_lenc1.describe()
sk=skew(sep_lenc1)
ku=kurtosis(sep_lenc1)
print(desc)
print(sk)
print(ku)
```

```python
sep_lenc2.describe()
sk=skew(sep_lenc2)
ku=kurtosis(sep_lenc2)
print(desc)
print(sk)
print(ku)
```

```python
sep_lenc3.describe()
sk=skew(sep_lenc3)
ku=kurtosis(sep_lenc3)
print(desc)
print(sk)
print(ku)
```
